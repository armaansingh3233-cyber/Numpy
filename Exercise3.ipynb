#### Numpy array using randint

```python
import numpy as np
electricity=np.random.randint(120,650,size=(4,7))
print(electricity)
print(electricity.shape)
print(electricity.size)


```

#### Average usuage per hostel,maximum usuage per hostel and highest average usuage 

```python
average_usuageperhostel=electricity.mean(axis=1)
print(average_usuageperhostel)

max_usuage=electricity.max(axis=1)
print(max_usuage)

highest_average_index=np.argmax(average_usuageperhostel)
print(highest_average_index)
```

#### Total Electricity Consumption

```python
total_consumption=electricity.sum()
print(total_consumption)
```

#### Broadcasting 

```python
new_electricity=np.where(electricity>550,electricity*1.20,electricity).astype(int)
print(new_electricity)
```

#### Comparing average hostel consumption to daily consumption

```python
compare_consumption = np.where(
    electricity > average_usuageperhostel[:, None],
    True,
    False
)

print(compare_consumption)
```

#### Reshaping

```python
print(electricity.reshape(1,28))
print(electricity.reshape(2,14))
```

#### Copy vs View

```python
ref_backup=electricity
safe_backup=electricity.copy()
electricity[2][3]=222
print(ref_backup)
print(safe_backup)
```
