```python
import numpy as np
marks=np.array([[78,85,90],[45,60,55],[88,92,95],[30,40,35],[65,70,75]])
```

#### Total Elements,Size and Data Type

```python
print(marks.size)
print(marks)
print(marks.dtype)
print(marks.shape)
```

#### Data Indexing and Slicing

```python
print(marks[1][:])
print(marks[:,2])
```

#### Mean and Average

```python
print(np.mean(marks,axis=1))
print(np.mean(marks,axis=0))
```

#### Filtering Top Performers

```python
avg_per_student = np.mean(marks, axis=1)

# students with average >= 80
result = marks[avg_per_student >= 80]
print(result)
```

#### Pass/fail

```python
result = np.where(marks >= 50, "Pass", "Fail")
print(result)
```

#### Data Cleaning

```python
updated_marks = np.where(marks < 40, 0, marks)
print(updated_marks)
```

#### Nested Condition


```python
avg = np.mean(marks, axis=1)

grades = np.where(avg >= 85, "A",
         np.where(avg >= 70, "B",
         np.where(avg >= 50, "C", "No Grade")))

scores = np.where(avg >= 85, 10,
         np.where(avg >= 70, 8,
         np.where(avg >= 50, 6, 0)))

print("Average:", avg)
print("Grade:", grades)
print("Score:", scores)
```

#### .copy vs =

```python
backup=marks
backup[0,0]=69
print(backup)
print(marks)

backup2=marks.copy()
backup2[0][0]=6969696996
print(backup2)
print(marks)
```
