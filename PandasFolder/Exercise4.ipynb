```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data={
    "Monthly_Inc":[15000,18000,20000,22000,17000,19000,21000,23000,16000,20000,18000,22000,21000,24000,17000,19000,20000,23000,18000,21000],
    "Online_Ord":[3,4,5,6,4,5,6,7,3,5,4,6,5,7,4,5,6,7,4,5],
    "Spotify_Hrs":[1,2,3,4,2,3,4,5,1,3,2,4,3,5,2,3,4,5,2,3],
    "Shop_Hours":[2,3,4,5,3,4,5,6,2,4,3,5,4,6,3,4,5,6,3,4],
    "Money_Spent":[4000,4500,5000,6000,4200,4800,5500,6500,4100,5000,4600,6000,5300,6800,4300,4900,5200,6700,4500,5400]
}

df=pd.DataFrame(data)

df
```

```python
print(df.mean())
```

```python
print(df.median())
```

```python

print(df.std())
```

```python
print(df.corr())
```

```python
plt.scatter(df["Monthly_Inc"], df["Money_Spent"])

plt.xlabel("Monthly Income")
plt.ylabel("Money Spent")
plt.title("Income vs Spending")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.hist(df["Online_Ord"], bins=5, edgecolor="black")

plt.xlabel("Number of Online Orders")
plt.ylabel("Frequency")
plt.title("Histogram of Online Orders")

plt.grid(axis="y")

plt.show()
```

```python
plt.figure(figsize=(6,5))

plt.plot(df["Spotify_Hrs"], marker='o')

plt.xlabel("User Index")
plt.ylabel("Spotify Hours")
plt.title("Spotify Listening Hours Trend")

plt.grid(True)

plt.show()
```

### Inference

#### Income vs Money Spent (Scatter Plot)

--More income,more the money spent.


#### Histogram of online orders

--Most individuals place moderate numbers of online orders.
--Very few people order extremely low or extremely high amounts.
--The distribution is centered around the middle values.

#### Spotify Hours Line Plot

--Users typically spend 1–5 hours on Spotify.
--Most people listen around 2–4 hours, indicating moderate music usage.

#### Correlation Analysis

--Monthly Income ↔ Money Spent: Strong positive correlation
--Online Orders ↔ Money Spent: People who order more tend to spend more
--Shopping Hours ↔ Money Spent: More shopping time leads to higher spending

#### Mean, Median, Standard Deviation

--Mean and Median are close → data is fairly balanced.
--Standard deviation is moderate → values are not extremely spread out.
--Spending behavior is consistent among most users.
