#### 1D Series

```python
import pandas as pd
import numpy as np
s=pd.Series([10,20,30,40,50])
print(s)
```




#### Custom Index


```python
s1=pd.Series([10,20,30],index=["a","b","c"])
print(s1)
print(s1["a"])
print(s[0])
```

#### Dictionary

```python
data={"Math":90,"Physics":55,"Chemistry":89}
print(data)
#Dictionary to series
marks=pd.Series(data)
print(marks)
```

#### Data Frame

```python
data={
    "Name":["Aman","Krish","Karan"],
    "Age":[22,23,21],
    "Marks":[85,90,88]
}
print(data)
print(type(data))

marks=pd.DataFrame(data)
print(marks)
```
