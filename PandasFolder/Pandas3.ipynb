#### Data Visualization

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df={
    "Name":["Aman","Riya","Rahul","Sneha","Arjun","Priya","Karan","Neha"],
    "Age":[22,25,24,23,26,27,28,22],
    "Gender":["Male","Female","Male","Female","Male","Female","Male","Female"],
    "Marks":[75,88,92,70,85,60,80,66]
}
data=pd.DataFrame(df)
```

```python
print(data)
```

```python
# We plotting one feature so this is univalent analysis
plt.plot(data["Marks"])
plt.title("Marks trend")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

```python
#Now we plot 2 features so bivalent analysis
plt.bar(data["Name"],data["Marks"])
plt.title("Marks of students")
plt.xlabel("Student name")
plt.ylabel("Marks")
plt.show()
```

```python
plt.hist(data["Age"],bins=5)
plt.title("Age distribution")
plt.xlabel("Age")
plt.ylabel("Age")
plt.show()
```

```python
plt.scatter(data["Age"],data["Marks"])
plt.title("Age vs Marks")
plt.xlabel("Age")
plt.ylabel("Marks")
plt.show()
```

```python
gender_counts = data["Gender"].value_counts()

plt.pie(gender_counts, labels=gender_counts.index, autopct="%1.1f%%")
plt.title("Gender Distribution")
plt.show()
```

```python
# Two graphs together
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.hist(data["Age"])
plt.title("Age Distribution")
plt.subplot(1,2,2)
plt.bar(data["Name"],data["Marks"])
plt.title("Marks of Students")
plt.show()
```
